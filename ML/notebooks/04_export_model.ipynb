```python
import joblib
from micromlgen import port

# 1. Load your trained model
clf = joblib.load("../../ML/models/model_v1.pkl")

# 2. Convert model to C code
c_code = port(clf)

# 3. Save as header file
with open("../../ML/exports/model_v1.h", "w") as f:
    f.write(c_code)

print("✅ Exported model to ML/exports/model_v1.h")
```

```python
plan_keys = [
    "C_CONS_3D_20_30",
    "C_MOD_3_4D_30_40",
    "C_STD_4D_35_45",
    "S_CONS_2_3D_30_40",
    "S_MOD_3D_35_50",
    "S_STD_3D_45_60",
]

with open("../../ML/exports/lookup_table.h", "w") as f:
    f.write("const char* PLAN_KEY_TABLE[] = {\n")
    for key in plan_keys:
        f.write(f'    "{key}",\n')
    f.write("};\n")

print("✅ Exported lookup table to ML/exports/lookup_table.h")
```
